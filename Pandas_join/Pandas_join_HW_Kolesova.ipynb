```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
active = ratings.groupby('userId')
```

```python
previous_user = None
timestamp = []

with open('ratings.csv') as f:
    f.readline()
    
    for line in f:
        line = line.strip().split(',')

        current_user_id = line[0]
        time = int(line[3])
        
        if previous_user:
            if previous_user == current_user_id:
                timestamp.append(time)
            else:
                if len(timestamp) >= 100:
                    print(previous_user, max(timestamp) - min(timestamp))
                
                previous_user = current_user_id
                timestamp = [time]
        else:
            previous_user = current_user_id
            timestamp.append(time)
```
