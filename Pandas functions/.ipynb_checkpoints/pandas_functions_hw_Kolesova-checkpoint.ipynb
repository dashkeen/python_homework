```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
class_rating = ''

def film_rating(param):
    """Определяет рейтинг как низкий, средний или высокий"""
    if param.rating <= 2.0:
        class_rating = 'low rated'
    elif 4.0 >= param.rating > 2.0:
        class_rating = 'average'
    elif param.rating >= 4.5:
        class_rating = 'high rated'
    return class_rating

data['class'] = data.apply(film_rating, axis = 1)
data.head(15)
```

```python
df = pd.read_csv('keywords.csv')

def region(row):
    """Определяет приинадлежность запроса какому-либо региону, если имеется упоминание"""
    reg = 'undefined'
    geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    for key, value in geo_data.items():
        for city in value:
            if city in row.keyword:
                reg = key
    return reg

df['region'] = df.apply(region, axis = 1)
df.loc[21440:21450]
```

```python
df1 = pd.read_csv("ratings.csv")
df2 = pd.read_csv("movies.csv")

df3 = pd.merge(df1, df2, on=["movieId"])
df3.to_csv("final.csv",index=False)

years = list(range(1950, 2010))

def prod_year(row):
    god = '1900'
    for year in years:
        if str(year) in row['title']:
            god = year
    return god
df3['year'] = df3.apply(prod_year, axis = 1)
```

```python
df4 = df3.groupby('year')
df4.rating.mean().sort_values(ascending=False).head(50)
```
